```python
import pandas as pd 

df_raw = pd.read_csv('UnifiedDataset.csv')
df_raw.sample(20)
```

```python
columns = df_raw.columns.to_list()
columns
```

```python
df = df_raw[['Country', 'Year', 'Infant Mortality Rate', 'Suicides Rate', '% Population Aged 0-14','% Population Aged 15-64','% Population Aged 65+','% Population Aged 65-69','% Population Aged 70-74','% Population Aged 75-79','% Population Aged 80+','Maternal Mortality Ratio',  'Neonatal Mortality Rate','Universal Heath Care Coverage','Birth Rate','Death Rate','GDP per Capita','% Population $1.90 a day','% Population $3.20 a day','% Population $5.50 a day','Income per Capita','Total Population','GNI per Capita','Conflict and Terrorism Deaths','Homicide Rate','Government Expenditure Education','Government Expenditure Military','Government Expenditure Health', ]]
```

```python
df.info()
```

```python
df = df.round(2)
df.sample(20)
```

### Exploratory Data Analysis ###

```python
df.describe().round(2)
```

```python
df_population = df.sort_values(by ='Total Population', ascending= False)
df_population
```

```python
df_2000 = df[df['Country']!= 'World']
df_2000 = df[df['Year']== 2000]
df_sorted = df_2000.sort_values(by='Total Population', ascending=False)
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(8,5))
plt.bar(df_sorted['Country'], df_sorted['Total Population']/1e6)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Countries by Population in 2000')
plt.show()
```

```python
df_Mexico = df[df['Country'] == 'Mexico'] 
df_Mexico = df_Mexico.sort_values(by = 'Total Population', ascending= False)
```

```python
plt.Figure(figsize=(8,5))
plt.bar(df_Mexico['Year'], df_Mexico['Total Population']/1e6)
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Mexico total population 1990-2019')
plt.show()
```
